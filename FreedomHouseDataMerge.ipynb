```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\jackm\Documents\Data Analytics\CW\ReligionMergeData.csv')
```

```python
dfFreedomHouse = pd.read_excel(r'C:\Users\jackm\Documents\Data Analytics\CW\FreedomHouseData.xlsx')
```

```python
dfFreedomHouse.head()
```

```python
#Drop first row and empty columns
dfFreedomHouse.drop(dfFreedomHouse.index[0], inplace = True)
dfFreedomHouse.drop(columns=['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 6', 'Unnamed: 7'], inplace =True)
```

```python
dfFreedomHouse.drop(dfFreedomHouse.index[0], inplace = True)
dfFreedomHouse.head()
```

```python
#rename columns, survey edition column not changing, possibly because is key
dfFreedomHouse.rename(columns={ 'Unnamed: 4': 'Civil Liberties'}, inplace = True)
dfFreedomHouse.rename(columns={ 'Unnamed: 5': 'Status'}, inplace = True)
dfFreedomHouse.rename(columns={ 'Survey Edition': 'Country'}, inplace = True)
dfFreedomHouse.rename(columns={ '2017': 'Personal Liberties'}, inplace = True)
```

```python
dfFreedomHouse.head()
```

```python
#remove white space and make all caps
dfFreedomHouse['Country'] = dfFreedomHouse['Country'].str.replace(" ", "")
dfFreedomHouse['Country'] = dfFreedomHouse['Country'].str.upper()
```

```python
dfFreedomHouse.head()
```

```python
#first merge test for null values
df_nullTest = pd.merge(df, dfFreedomHouse, on ='Country', how='left')
```

```python
allnulls_data = df_nullTest[df_nullTest.isnull().any(axis=1)]
```

```python
print(allnulls_data[allnulls_data["Civil Liberties"].isnull()])
```

```python
#String search on nulls: Hong Kong
dfFreedomHouse[dfFreedomHouse['Country'].str.contains("HONG")]
#Hong Kong not listed in freedomhouse data
```

```python
#Search 
dfFreedomHouse[dfFreedomHouse['Country'].str.contains("MACEDONIA")]
```

```python
#rename macedonia
dfFreedomHouse['Country'] = dfFreedomHouse['Country'].replace(['NORTHMACEDONIA'], ['MACEDONIA'])
```

```python
#Search Palestine
dfFreedomHouse[dfFreedomHouse['Country'].str.contains("PALES")]
#not included
```

```python
#Remerge with amendments
dfFinal = pd.merge(df, dfFreedomHouse, on ='Country', how='left')
```

```python
dfFinal.drop(columns = 'Unnamed: 0', inplace = True)
```
