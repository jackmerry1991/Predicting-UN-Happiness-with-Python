```python
from sklearn.linear_model import Ridge
import seaborn as sns; sns.set()
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#load data
df = pd.read_csv(r'C:\Users\jackm\Documents\Data Analytics\CW\FINALMERGE.csv')
df.drop(columns='Unnamed: 0', inplace = True)
```

```python
df.head()
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(df.drop(columns = 'Happiness.Score'), df['Happiness.Score'], test_size = 0.2, random_state = 0)
xTrain.head()
```

```python
from sklearn.linear_model import Ridge
from math import sqrt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

rr = Ridge(alpha=0.01)
rr.fit(xTrain, yTrain) 
pred_train_rr= rr.predict(xTrain)

print(np.sqrt(mean_squared_error(yTrain,pred_train_rr)))
print(r2_score(yTrain, pred_train_rr))

pred_test_rr= rr.predict(xTest)
print(np.sqrt(mean_squared_error(yTest,pred_test_rr))) 
print(r2_score(yTest, pred_test_rr))
```

```python
target_column = df['Happiness.Score'] 
predictors = df.drop(columns = 'Happiness.Score')
predictors.head()
```

```python
X = predictors.values
y = target_column.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)
print(X_train.shape); print(X_test.shape)
```

```python
#linear regression model
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#first line predicts on training set, second and third print evaluation metrics, then repeated on test set lines 4-6

pred_train_lr= lr.predict(X_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_lr)))
print(r2_score(y_train, pred_train_lr))

pred_test_lr= lr.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred_test_lr))) 
print(r2_score(y_test, pred_test_lr))
```

```python
# ridge regression again
rr = Ridge(alpha=0.01)
rr.fit(X_train, y_train) 
pred_train_rr= rr.predict(X_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_rr)))
print(r2_score(y_train, pred_train_rr))

pred_test_rr= rr.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred_test_rr))) 
print(r2_score(y_test, pred_test_rr))
```

```python
# lasso regression
from sklearn.linear_model import Lasso

model_lasso = Lasso(alpha=0.01)
model_lasso.fit(X_train, y_train) 
pred_train_lasso= model_lasso.predict(X_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_lasso)))
print(r2_score(y_train, pred_train_lasso))

pred_test_lasso= model_lasso.predict(X_test)
print(np.sqrt(mean_squared_error(y_test,pred_test_lasso))) 
print(r2_score(y_test, pred_test_lasso))
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(X_test)
# Calculate the absolute errors
errors = abs(predictions - y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2))
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
