```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv(r'C:\Users\jackm\Documents\Data Analytics\CW\PreProcessingOne.csv')
```

```python
df.head()
```

```python
df.drop(columns= 'Unnamed: 0', inplace=True)
```

```python
#load religion data
dfReligion = pd.read_csv(r'C:\Users\jackm\Documents\Data Analytics\CW\ReligionData.csv')
```

```python
dfReligion.head()
```

```python
#Rename columns
dfReligion.rename(columns={ 'Unnamed: 3': 'Religion'}, inplace = True)
dfReligion.rename(columns={ 'Entity': 'Country'}, inplace = True)
```

```python
dfReligion.head()
```

```python
#Drop blank Columns
dfReligion.drop(['Code', 'Year', 'Unnamed: 4', 'Unnamed: 5'], axis = 1, inplace = True)
```

```python
dfReligion.head()
```

```python
#edit country names to make compatible with main dataframe
dfReligion['Country'] = dfReligion['Country'].str.replace(" ", "")
dfReligion['Country'] = dfReligion['Country'].str.upper()
```

```python
#outer join to check where names are not compatible
df_nullTest = pd.merge(df, dfReligion, on ='Country', how='left')
```

```python
#sum total of null to find those that didn't join
df_nullTest.isnull().sum()
```

```python
df_nullTest.head()
```

```python
#create new df of only the null values
allnulls_data = df_nullTest[df_nullTest.isnull().any(axis=1)]
```

```python
print(allnulls_data[allnulls_data["Religion"].isnull()]) 
# 4 missing
```

```python
#string search on those missing Hong Kong
dfReligion[dfReligion['Country'].str.contains("HONG")]
```

```python
# Change Hong Kong to match
dfReligion['Country'] = dfReligion['Country'].replace(['HONGKONG'], ['HONGKONGS.A.R.,CHINA'])
```

```python
# Search Palestine
dfReligion[dfReligion['Country'].str.contains("PALESTINE")]
```

```python
# Change Palestine to match
dfReligion['Country'] = dfReligion['Country'].replace(['HONGKONG'], ['PALESTINIANTERRITORIES'])
```

```python
# Congo Search
dfReligion[dfReligion['Country'].str.contains("CONGO")]
```

```python
# Change Congo
dfReligion['Country'] = dfReligion['Country'].replace(['CONGO'], ['CONGO(BRAZZAVILLE)'])
dfReligion['Country'] = dfReligion['Country'].replace(['DEMOCRATICREPUBLICOFCONGO'], ['CONGO(KINSHASA)'])
```

```python
# Merge again with alterations
dfFinal = pd.merge(df, dfReligion, right_index=True
```
